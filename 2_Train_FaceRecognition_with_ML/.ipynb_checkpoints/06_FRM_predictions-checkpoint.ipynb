```python
# cv2.imshow("Predicted image",pred_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()


from matplotlib import pyplot as plt

# Call the updated function
pred_img, predictions = faceRecognitionPipeline('test_images/getty_test.jpg')

print(f"Detected {len(predictions)} face(s)")

# Display in Jupyter
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB))
plt.title(f"Face Recognition Results - {len(predictions)} faces detected")
plt.axis('off')
plt.show()

# Print predictions
for i, pred in enumerate(predictions):
    print(f"Face {i+1}: {pred['prediction_name']} (confidence: {pred['score']*100:.2f}%)")
```

```python
img_rgb = cv2.cvtColor(pred_img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()

## generate report
for i in range(len(pred_dict)):
    obj_gray = pred_dict[i]['roi'] #gray scale
    obj_eig = pred_dict[i]['eig_img'].reshape(100,100) #eigen image
    plt.subplot(1,2,1)
    plt.imshow(obj_gray,cmap='gray')
    plt.title('Gray ScaleImage')
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(obj_eig,cmap='gray')
    plt.title('Eigen Image')
    plt.axis('off')
    
    plt.show()
    print('Predicted Gender =',pred_dict[i]['prediction_name'])
    print('Predicted score = {:,.2f} %'.format(pred_dict[i]['score']*100))
    
    print('-'*100)
```

```python
# cap = cv2.VideoCapture('./test_images/video.mp4') # 0 -> access webcamera, 1-> external camera, path

# while True:
#     ret, frame = cap.read()
    
#     if ret == False:
#         break
    
#     pred_img, pred_dict = faceRecognitionPipeline(frame,path=False)
    
#     cv2.imshow('prediction',pred_img)
#     if cv2.waitKey(1) == ord('q'):
#         break
    
# cap.release()
# cv2.destroyAllWindows()



import cv2
from tqdm.notebook import tqdm

cap = cv2.VideoCapture('./test_images/video.mp4')

# Get total frames for progress bar
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Setup output video
output_path = 'output_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Process video
for _ in tqdm(range(total_frames), desc="Processing video"):
    ret, frame = cap.read()
    
    if ret == False:
        break
    
    pred_img, pred_dict = faceRecognitionPipeline(frame, path=False)
    out.write(pred_img)

cap.release()
out.release()
print("Video processing complete")

# Display the result
from IPython.display import Video
Video(output_path, width=800)
```
