```python
import numpy as np
import cv2
import matplotlib.image as matimg
from PIL import Image

import matplotlib.pyplot as plt
```

```python
# OpenCV - BGR - numpy array
img_cv = cv2.imread('data/flying_bird.jpg')
# Matplotlib - RGB - numpy array
img_mat = matimg.imread('data/flying_bird.jpg')
# PIL - RGB - object
img_pil = Image.open('data/flying_bird.jpg')
```

### Display Image

```python
# display image using opencv
import cv2
from matplotlib import pyplot as plt

# Convert BGR to RGB for correct colors
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')  # Optional: hide the x and y axes
plt.show()
```

```python
# display image using matplotlib
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1,3,3)
plt.imshow(img_pil_arr)
plt.axis('off')
plt.title('Pillow RGB')

plt.show()
```

### COLOR Space

- Color Conversion
- Gray Scale

```python
img_cv.shape
```

```python
# numpy slicing
# BGR
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
import cv2
import matplotlib.pyplot as plt

# 1. Convert the original BGR image to RGB for Matplotlib
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

# 2. Setup the subplot (1 row, 4 columns)
plt.figure(figsize=(20, 5))

# Original Image
plt.subplot(1, 4, 1)
plt.imshow(img_rgb)
plt.title('BGR (Converted to RGB)')
plt.axis('off')

# Blue Channel (Displayed as Grayscale)
plt.subplot(1, 4, 2)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

# Green Channel (Displayed as Grayscale)
plt.subplot(1, 4, 3)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

# Red Channel (Displayed as Grayscale)
plt.subplot(1, 4, 4)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.show()

```

```python
# opencv split channels
b,g,r  = cv2.split(img_cv)
```

```python
import matplotlib.pyplot as plt
import cv2

# Convert original BGR image to RGB so colors look correct in the notebook
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

# Create a figure to hold the 4 images
plt.figure(figsize=(20, 5))

# 1. Show Blue Channel (as grayscale)
plt.subplot(1, 4, 1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

# 2. Show Green Channel (as grayscale)
plt.subplot(1, 4, 2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

# 3. Show Red Channel (as grayscale)
plt.subplot(1, 4, 3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

# 4. Show Original Image (converted to RGB)
plt.subplot(1, 4, 4)
plt.imshow(img_rgb)
plt.title('Original BGR')
plt.axis('off')

plt.tight_layout()
plt.show()
```
