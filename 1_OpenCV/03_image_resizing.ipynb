```python
import cv2
import matplotlib.pyplot as plt

# 1. Loading image
img = cv2.imread('data/flying_bird.jpg')

# 2. Print shape (Standard OpenCV)
print(f"Image Shape: {img.shape}")

# 3. Convert BGR to RGB
# (OpenCV loads as BGR, but Matplotlib needs RGB to show colors correctly)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 4. Show inside Notebook using Matplotlib
plt.imshow(img_rgb)
plt.title('Original')
plt.axis('off')  # Hides pixel coordinates
plt.show()
```

###  Task
1. Shink the image
2. Enlarge the image

```python
# shinking image
# width = columns , height = rows
img_shink = cv2.resize(img,dsize=(320,240),interpolation=cv2.INTER_AREA)

# enlarge 
# cv2.INTER_CUBIC, cv2.INTER_LINEAR
img_zoom = cv2.resize(img,dsize=(1280,960),interpolation=cv2.INTER_CUBIC) # 
```

```python
img_shink.shape, img_zoom.shape
```

```python
import matplotlib.pyplot as plt
import cv2

# 1. Convert BGR to RGB for each image so colors display correctly
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_shrink_rgb = cv2.cvtColor(img_shink, cv2.COLOR_BGR2RGB) # Note: matching your variable 'img_shink'
img_zoom_rgb = cv2.cvtColor(img_zoom, cv2.COLOR_BGR2RGB)

# 2. Create a figure to display all three side-by-side
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title(f'Original\n{img.shape[:2]}')
plt.axis('off')

# Shrinked Image
plt.subplot(1, 3, 2)
plt.imshow(img_shrink_rgb)
plt.title(f'Shrink\n{img_shink.shape[:2]}')
plt.axis('off')

# Zoomed Image
plt.subplot(1, 3, 3)
plt.imshow(img_zoom_rgb)
plt.title(f'Zoom\n{img_zoom.shape[:2]}')
plt.axis('off')

plt.tight_layout()
plt.show()
```
